# Primeros pasos con NumPy y Pandas

```python
# Librería
import numpy as np

# Semilla para pseudo-aleatorios
np.random.seed(1)

# Creación de diccionario
d = {"A" : np.random.binomial(10, 0.5, 5),
     "B" : np.random.binomial(10, 0.3, 5),
     "C" : np.random.binomial(10, 0.6, 5),
     "D" : np.random.binomial(10, 0.4, 5)}

d
```

```python
# Importación de la Librería
import pandas as pd

# Convertir diccionario en Dataframe
df = pd.DataFrame(d)
df
```

```python
idx = ["A", "B", "C", "D", "E"]

df = pd.DataFrame(d, index=idx)
df["Promedio"] = df.mean(axis=1)
df
```

```python
# Obtener nombres de columnas
df.columns
```

```python
# Escoger columnas
df[['A', 'C']]
```

```python
# Obtener nombre de las columnas (variables) del dataframe
ssma.columns
```

```python
# Creación de nuevo dataframe según determinadas columnas
ssma_1 = ssma[['Age', 'Gender', 'Avg_Daily_Usage_Hours', 'Most_Used_Platform']]
ssma_1
```

```python
# Filtrar dataframe con respecto a la condición de una columna
avg = ssma_1[ssma_1['Avg_Daily_Usage_Hours'] > 4]
avg
```

```python
ssma_1 = ssma_1.rename(columns={'Avg_Daily_Usage_Hours':'Hours','Most_Used_Platform':'Platform'})
ssma_1
```

```python
ssma_2 = ssma_1[ssma_1['Hours'] > 4]
ssma_2
```

```python
ssma_2.info()
```

```python
group_ssma_2 = ssma_2.groupby(by = ['Gender'])['Hours'].mean()
group_ssma_2
```

```python
#platform_ssma2 = ssma_2.groupby(by=['Platform'])['Hours'].mean().sort_values(ascending=False)
#platform_ssma2 = ssma_2.groupby(by=['Platform'])['Hours'].max().sort_values(ascending=False)
platform_ssma2 = ssma_2.groupby(by=['Platform'])['Hours'].count().sort_values(ascending=False)
platform_ssma2
```

```python
ssma_3 = ssma_2.groupby('Platform')['Hours'].size().sort_values(ascending=False)
ssma_3
```
