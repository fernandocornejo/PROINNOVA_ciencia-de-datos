# Primeros pasos con NumPy y Pandas

```python
# Librería
import numpy as np

# Semilla para pseudo-aleatorios
np.random.seed(1)

# Creación de diccionario
d = {"A" : np.random.binomial(10, 0.5, 5),
     "B" : np.random.binomial(10, 0.3, 5),
     "C" : np.random.binomial(10, 0.6, 5),
     "D" : np.random.binomial(10, 0.4, 5)}

d
```

```python
# Importación de la Librería
import pandas as pd

# Convertir diccionario en Dataframe
df = pd.DataFrame(d)
df
```

```python
# Obtener nombres de columnas
df.columns
```

```python
# Escoger columnas
df[['A', 'C']]
```

```python
# Escoger filas
df.iloc[::2]
```

```python
# Estructura del dataframe
df.info()
```

```python
#Estadística descriptiva de los datos
df.describe()
```

```python
#Transponer
df.T
```

```python
#Selección específica
print(df['A'][0])
```
