#### Función `range()`
Es un generador de índices

```python
range(1,10)
```

Para visualizar los elementos de un generador, se necesitan bules `for`

```python
for i in range(1,11):
  print(i)
```

```python
L2 = [i for i in range(11)]
L2
```

## Condicionales
- if, else, else if (elif)
- while

```python
L3 = [i for i in range(11) if i%2 == 0]
L3
```
